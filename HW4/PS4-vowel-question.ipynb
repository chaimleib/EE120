```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Graphing helper function
def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

Before running the cell below, capture two short audio files of the vowels "eeee" and "oooo".
Audacity works nicely with File -> Export Selection to WAV signed 16 bit PCM.

```python
# import file
from scipy.io import wavfile
rate1,data1= wavfile.read('vowel-e.wav')  # 16 bit data if from Audacity
rate1,data2= wavfile.read('vowel-o.wav')  # 16 bit data if from Audacity
print 'rate1 =', rate1
print 'data1 =', data1
lengthE = np.size(data1)/2
print 'length E = ', lengthE
t1 = np.linspace(0, float(lengthE)/float(rate1), lengthE)
#
print 'data2 =', data2
lengthO = np.size(data2)/2
print 'length O = ', lengthO
t2 = np.linspace(0, float(lengthO)/float(rate1), lengthO)
```

In the cell below, select the period (number of samples) for the 2 vowels. This can be determined approximately by inspection of the waveform.

```python
# plot data
# t1 = np.linspace(0, float(length)/float(rate1), length)\
length = max(lengthE, lengthO)
vowelE = np.zeros(length)
vowelO = np.zeros(length)
for i in range(0,length):
    if i < lengthE:
        vowelE[i] = data1[i][0] # copy left channel
    
    if i < lengthO:
        vowelO[i] = data2[i][1]  # copy left channel
# Choose N 
Ne = 385 # estimated period of vowel E  TO BE DETERMINED
No = 371 # estimated period of vowel O  TO BE DETERMINED
_ = plt.plot(range(0,2*Ne), vowelE[0:2*Ne])
_ = plt.plot(range(0,2*No), vowelO[0:2*No])
```

Select appropriate k for reasonable approximation.

```python
# calculate Fourier Series
kmax = 35
coeffE = np.exp(1j* np.zeros(kmax))  # a_k. Since real signal, can use conjugate symmetry for a_{-k}
omegaE = 2.0 * np.pi/Ne
sum = complex(0,0)  # initialize to complex
for k in range(0,kmax):   # pick number of Fourier Series coefficients
    sum = 0.0+ 0j
    for n in range(0,Ne):
            sum = sum + vowelE[n] * np.exp(-1j * omegaE * n * k)          
    coeffE[k] = sum/Ne
print 'coefficients 0-7:'
#print "real:", np.real(coeffE[0:8])
#print "imag:", np.imag(coeffE[0:8])
from pprint import pprint
pprint(list(np.real(coeffE[0:8])), width=1)
pprint(list(np.imag(coeffE[0:8])), width=1)
setup_graph(title='$O$', x_label='$k$', y_label='real $a_k$', fig_size=(6,3))
_ = plt.stem(range(0,kmax), np.real(coeffE))
setup_graph(title='$O$', x_label='$k$', y_label='imag $a_k$', fig_size=(6,3))
_ = plt.stem(range(0,kmax), np.imag(coeffE))
```

```python
# calculate Fourier Series
coeffO = np.exp(1j* np.zeros(kmax))  # a_k. Since real signal, can use conjugate symmetry for a_{-k}
omegaO = 2.0 * np.pi/No
sum = complex(0,0)  # initialize to complex
for k in range(0,kmax):   # pick number of Fourier Series coefficients
    sum = 0.0+ 0j
    for n in range(0,No):
            sum = sum + vowelO[n] * np.exp(-1j * omegaO * n * k)          
    coeffO[k] = sum/No
print 'coefficients 0-7:'
#print "real:", np.real(coeffO[0:8])
#print "imag:", np.imag(coeffO[0:8])
pprint(list(np.real(coeffO[0:8])), width=1)
pprint(list(np.imag(coeffO[0:8])), width=1)
setup_graph(title='$O$', x_label='$k$', y_label='real $a_k$', fig_size=(6,3))
_ = plt.stem(range(0,kmax), np.real(coeffO))
setup_graph(title='$O$', x_label='$k$', y_label='imag $a_k$', fig_size=(6,3))
_ = plt.stem(range(0,kmax), np.imag(coeffO))
```

```python
# synthesize vowel using a_k coefficients
xe = np.zeros((Ne,2))#just repeat one period
xe_imag = np.zeros((Ne,2))#just repeat one period
for n in range(0,Ne):
    sum = 0+0j
    for k in range(0,kmax): # pick number of Fourier Series coefficients
            sum = sum + coeffE[k]*np.exp(1j * omegaE *n*k) + \
                    np.conjugate(coeffE[k]) * np.exp(-1j * omegaE * n * k)           
    xe[n,0]=np.real(sum) # left channel
    xe[n,1]=np.real(sum) # right channel
    xe_imag[n,0] = np.imag(sum)
setup_graph(title='$E_{%d}$' % kmax, x_label='$n$', y_label='x[n]', fig_size=(6,3))
_ = plt.plot(range(0,Ne), xe)
_ = plt.plot(range(0,Ne), vowelE[0:Ne])
#_ = plt.plot(range(0,Ne), xe_imag)  # plot to verify zero
```

```python
# synthesize vowel using a_k coefficients
xo = np.zeros((No,2))#just repeat one period
xo_imag = np.zeros((No,2))#just repeat one period
for n in range(0,No):
    sum = 0+0j
    for k in range(0,kmax): # pick number of Fourier Series coefficients
            sum = sum + coeffO[k]*np.exp(1j * omegaO *n*k) + \
                    np.conjugate(coeffO[k]) * np.exp(-1j * omegaO * n * k)           
    xo[n,0]=np.real(sum) # left channel
    xo[n,1]=np.real(sum) # right channel
    xo_imag[n,0] = np.imag(sum)
setup_graph(title='$O_{%d}$' % kmax, x_label='$n$', y_label='x[n]', fig_size=(6,3))
_ = plt.plot(range(0,No), xo)
_ = plt.plot(range(0,No), vowelO[0:No])

#_ = plt.plot(range(0,No), xo_imag)  # plot to verify zero
```

```python
# convert single period to full length
synthE = np.zeros((lengthE,2)) # initialize to zero
for i in range(0,lengthE):
    synthE[i] = xe[np.mod(i,Ne)]
_ = plt.plot(t1, synthE)
print 'synthE', synthE[0:8]
# now write  data file
wavfile.write('synthE.wav', rate1, synthE.astype(int16)) # 16 bit integer
setup_graph(title='$x[n]$', x_label='$n$', y_label='$ real x[n]$', fig_size=(6,3))
_ = plt.plot(t1[1000:1500],synthE[1000:1500])
```

```python
# convert single period to full length
synthO = np.zeros((lengthO,2)) # initialize to zero
for i in range(0,lengthO):
    synthO[i] = xo[np.mod(i,No)]
_ = plt.plot(t2, synthO)
print 'synthO', synthO[0:8]
# now write  data file
wavfile.write('synthO.wav', rate1, synthO.astype(int16)) # 16 bit integer
setup_graph(title='$x[n]$', x_label='$n$', y_label='$ real x[n]$', fig_size=(6,3))
_ = plt.plot(t2[1000:1500],synthO[1000:1500])
```
